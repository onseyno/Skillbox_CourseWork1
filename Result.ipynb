```python
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
import calendar

courses = pd.read_csv('data/courses.csv', usecols=[1,2,3], index_col=0)
# 15 строк, таблица без пропусков
course_contents = pd.read_csv('data/course_contents.csv', index_col=0)
# 1201 строка, таблица без пропусков
progresses = pd.read_csv('data/progresses.csv', index_col=0)
# 15934 строк, таблица без пропусков
progress_phases = pd.read_csv('data/progress_phases.csv', parse_dates=['start_date','finish_date'])
#progress_phases = pd.read_csv('data/progress_phases.csv')
# 350677 строк, таблица с 16169  строчками без finish_dates,
# это незаконченные на данный момент прогрессы
dataset = progress_phases.merge(progresses, how='left', left_on='progress_id', right_index=True)
dataset = dataset.merge(courses, how='left', left_on='course_id', right_index=True)
#dataset = dataset.merge(course_contents.loc[:, ~course_contents.columns.isin(['module_number', 'lesson_number'])], how='left', left_on='course_id', right_index=True)
dataset = dataset.merge(course_contents, how='left', left_on=['course_id','module_number','lesson_number'], right_on=[course_contents.index,'module_number','lesson_number'])
dataset.columns = ['progress_id','module_number','lesson_number','status','start_date','finish_date','student_id','course_id','course_title','course_field','module_title','lesson_title','lesson_token','is_video','is_homework']
dataset[dataset.student_id == '768c2987a744c51ce64a5993a2a94eaf'] #проверка прошла успешно
```

<b>1.1.6 Минимальный, максимальный, средний, медианный возраст студентов:</b>

```python
students = pd.read_csv('data/students.csv', usecols=[1,2,3], index_col=0)
# 43830 строк, у многих студентов пропущен город (city): 30668 строк и/или дата рождения (birthday): 18331 строка
# оставляем только birthday, чтобы не удалить случайно строки с путсыми городом:
students_bd_cleared = students['birthday'].copy()
students_bd_cleared = students_bd_cleared.dropna().reset_index()
#создаем колонку year, чтобы построить служебную гистограмму по десятилетиям и определить диапазон для фильтрации
students_bd_cleared['year'] = students_bd_cleared['birthday'].apply(lambda x: int(x[:4])//10)
students_bd_cleared.groupby('year').agg('count')['birthday']
```

```python
def func(string):
    if  (1960 < (int(string[:4])) < 2005) & (0 < int(string[5:7]) < 13) & (0 < int(string[8:]) < 32):
        return string
    else:
        return False  

#исключаем записи:
students_bd_cleared['birthday'] = students_bd_cleared['birthday'].apply(func)
students_bd_cleared = students_bd_cleared[students_bd_cleared['birthday'] != False]
#берем сегодняшнюю дату:
today = datetime.date.today()
data = datetime.datetime(today.year, today.month, today.day)
#считаем возраст и переводим в целые года:
students_bd_cleared = students_bd_cleared.astype({'birthday': 'datetime64'})
students_bd_cleared['age'] = (data - students_bd_cleared['birthday'])/1000000000/60/60/24/365
students_bd_cleared = students_bd_cleared.astype({'age': 'int'})
min_age = np.min(students_bd_cleared['age'])
max_age = np.max(students_bd_cleared['age'])
avg_age = np.mean(students_bd_cleared['age'])
med_age = np.median(students_bd_cleared['age'])
print(min_age, max_age, avg_age, med_age)
```

<b>4.1 Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение количества студентов, приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле), к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле).</b>

```python
dataset_hw_completed = dataset[(dataset['is_homework'] == True) & (dataset['status'] == 'done')].groupby(['course_title','module_number','student_id']).agg({'finish_date':'max'}).reset_index()
dataset_hw_next = dataset[(dataset['is_homework'] == True) & (dataset['module_number'] > 1)].groupby(['course_title','module_number','student_id']).agg({'start_date':'min'}).reset_index()
dataset_hw_next['module_number_prev'] = dataset_hw_next['module_number'].apply(lambda x: x-1)

dataset_hw_completed.to_csv('completed.csv')
dataset_hw_next.to_csv('next.csv')

dataset_hw_modules = dataset_hw_completed.groupby('course_title').agg('max')['module_number'].reset_index()
dataset_hw_modules.columns = ['course_title','maxmodule']
dataset_hw_completed = dataset_hw_completed.merge(dataset_hw_modules,how='left',on='course_title')
dataset_hw_completed = dataset_hw_completed[dataset_hw_completed['module_number'] != dataset_hw_completed['maxmodule']]

dataset_conversion = dataset_hw_completed.merge(dataset_hw_next[['course_title','student_id','module_number_prev','start_date']], how='left', left_on=['course_title','student_id','module_number'], right_on=['course_title','student_id','module_number_prev']).reset_index()

dataset_conversion2 = dataset_conversion.groupby(['course_title','module_number']).agg({'finish_date':'count','start_date':'count'})[['finish_date','start_date']].reset_index()
dataset_conversion2['conversion'] = dataset_conversion2['start_date'] / dataset_conversion2['finish_date']
#Не на всех модулях есть дз, а значит и конверсия может оказатсья нулевой, считаем такой вариант неправдоподбным и исключаем все строки
#в данных есть ошибки, связанные с "потерей" информации о дз студентов на некоторых модулях, что вызывает провалы в конверсии на некоторых модулях
i = 0
for i in range(len(dataset_conversion2)-1):
    if (dataset_conversion2.start_date[i] < dataset_conversion2.finish_date[i+1]) & (dataset_conversion2.course_title[i] == dataset_conversion2.course_title[i+1]):
        dataset_conversion2 = dataset_conversion2.drop(labels = [i],axis = 0)
        
dataset_conversion2 = dataset_conversion2[dataset_conversion2['conversion'] > 0]
dataset_conversion2
```

Для улучшения качества контента полезно выявить причину. На основании имеющихся данных придумайте метрику успеваемости студента. **Обоснуйте ее.** Выявите таких студентов на каждом курсе, чтобы предоставить продюсерам список проблемных студентов. 

Введем несколько критериев, совокупность которых будет характеризовть успеваемость студентов:
1. Количество модулей курса, на которые студент затратил больше времени, чем медианное время прохождение данного модуля всеми студентами. Отсортируем студентов по процентному кол-ву модулей в курсе, в которых они находятся в половине "задержавшихся". Те, у кого это больше 75 %, включим в список


```python
dataset51 = dataset[(dataset['is_homework'] == True) & (dataset['status'] == 'done') ]
dataset51['duration'] = (dataset51['finish_date'] - dataset51['start_date']).round('d')
dataset51.groupby(['course_title','module_number','student_id']).agg({'duration':'sum'}).reset_index()

dataset52 = dataset[(dataset['is_homework'] == True) & (dataset['status'] == 'done')].groupby(['course_title','student_id','module_number']).agg({'start_date':'min','finish_date':'max'}).reset_index()
dataset52['duration'] = (dataset52['finish_date'] - dataset52['start_date']).round('d')
dataset52 = dataset52.groupby(['course_title','module_number']).agg({'duration':'median'}).reset_index()

dataset53 = dataset.groupby('course_title').agg(lambda x: x.unique().size)['module_number'].reset_index()


dataset5 = dataset51.merge(dataset52, how='left', on=['course_title','module_number'], suffixes=['_ind','_med'])

dataset5['duration_flag'] = (dataset5['duration_ind'] > dataset5['duration_med']).astype(int) 
dataset5 = dataset5.groupby(['course_title','student_id']).agg({'duration_flag':'sum'}).sort_values(by='duration_flag', ascending=False).reset_index()
dataset5 = dataset5.merge(dataset53,how='left', on='course_title')
dataset5['ratio'] = dataset5['duration_flag'] / dataset5['module_number']
dataset53 = dataset5[dataset5['ratio'] > 0.75].sort_values(by='ratio', ascending=False)
dataset53
```
